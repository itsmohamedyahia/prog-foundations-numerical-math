### Cartesian coordinate system

```python
import matplotlib.pyplot as plt
import random as r
```

```python
x = 5
y = 7
plt.plot(x,y,'ro')
plt.show()
```

```python
for i in range(0,5):
    x = r.randint(-3,3)
    y = r.randint(-3,3)
    plt.plot(x,y,'o')
plt.axis('square')
plt.grid()
plt.show()
```

```python
ax = plt.axes()
ax.arrow(2,1,5,6, head_width=0.2, head_length=0.2)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
