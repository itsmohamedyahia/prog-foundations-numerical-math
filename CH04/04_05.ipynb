### Creating 3D shapes

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
axes = [10, 10, 10]
data = np.ones(axes)

cube = plt.figure()
ax = cube.add_subplot(projection='3d')

ax.voxels(data, facecolors='green', alpha=0.25)
plt.show()
```
